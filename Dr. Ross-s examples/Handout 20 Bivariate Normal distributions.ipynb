# Handout 20 Bivariate Normal distributions

```python
from symbulate import *
%matplotlib inline
```

## Exer 20.1

```python
M, R = RV(BivariateNormal(mean1 = 635, mean2 = 595, sd1 = 85, sd2 = 70, corr = 0.6))

mr = (M & R).sim(10000)
mr.plot()
```

```python
mr.plot(["density", "marginal"])
```

```python
m = M.sim(10000)
m.plot(bins = 30)
m.mean(), m.sd(), m.count_gt(700) / 10000
```

```python
m = (M | (abs(R - 700) < 5) ).sim(10000)
m.plot(bins = 30)
m.mean(), m.sd(), m.count_gt(700) / 10000
```

```python
d = (M - R).sim(10000)
d.plot(bins = 30)
d.mean(), d.sd(), d.count_gt(0) / 10000
```

## Ex 19.2

```python
X, I = RV(Normal(0, 1) * BoxModel([-1, 1]))

Y = I * X

y = Y.sim(10000)
y.plot(bins = 30)
y.mean(), y.sd()
```

```python
(X & Y).sim(10000).plot()
```

```python
(X & Y).sim(10000).plot(["scatter", "marginal"])
```

```python
xy = (X & Y).sim(10000)
xy.cov(), xy.corr()
```

```python
z = (X + Y).sim(10000)
z.plot(bins = 30)
z.count_eq(0) / 10000
```
