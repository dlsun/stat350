## Exer 21.2

```python
P = BoxModel([-1, 1], size=4)
Z = RV(P)
X = RandomProcess(P, TimeIndex(fs=1))

X[0] =  0
for n in range(4):
    X[n+1] = X[n] + Z[n]
```

```python
X.sim(1).plot(alpha=1)
```

```python
X.sim(3).plot(alpha=0.5)
```

```python
X.sim(100).plot()
```

```python
x = X[2].sim(10000)
x.plot()
x.mean(), x.var()
```

```python
xy = (X[2] & X[3]).sim(10000)
xy.plot(jitter=True, alpha=0.01)
xy.cov(), xy.corr()
```

```python
xy.plot("tile")
```

```python
xy.tabulate(normalize=True)
```

## Exer 21.3

```python
def f(outcome, time):
    return outcome * cos(2 * pi * time)

P = BoxModel([0.5, 1, 2])

X = RandomProcess(P, TimeIndex(fs = inf), f)
```

```python
X.sim(1).plot(alpha = 1, tmax = 3)
```

```python
X.sim(100).plot(tmax = 3)
```

```python
X[1].sim(10000).plot()
```

```python
X[2/3].sim(10000).plot()
```

```python
X[0.25].sim(10000).plot()
```

```python
def f(outcome, t):
    return outcome[0] * t + outcome[1]

X = RandomProcess(BoxModel([0, 1, 2]) * BoxModel([0, 1], probs=[0.8, 0.2]), TimeIndex(fs=inf), f)
```

```python
X.sim(1000).plot(alpha=0.01)
```

```python
x = X[1].sim(10000)
x.plot(),
x.mean(), x.var()
```

```python
x = X[2].sim(10000)
x.plot(),
x.mean(), x.var()
```

```python
xy = (X[1] & X[2]).sim(10000)
xy.plot(["tile", "marginal"])
xy.cov()
```

```python
xy.tabulate(normalize=True)
```
