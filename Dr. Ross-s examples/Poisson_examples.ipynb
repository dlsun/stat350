## Some Poisson distributions

```python
RV(Poisson(3.2)).sim(10000).plot()
Poisson(3.2).plot()
```

```python
for lam in [1.5, 4.2, 10]:
    RV(Poisson(lam)).sim(10000).plot(jitter = True)
    Poisson(lam).plot()
```

## Exer 10.1

```python
X = RV(Poisson(0.3))
x = X.sim(10000)
x.plot()
Poisson(0.3).plot()
x.tabulate(normalize=True)
```

```python
x.count_eq(0) / 10000
```

```python
x.count_eq(2) / 10000
```

```python
x.count_geq(3) / 10000
```

```python
X, Y = RV(Poisson(0.3) * Poisson(0.3))
(X+Y).sim(10000).plot()
Poisson(0.3 + 0.3).plot()
```

```python
Poisson(0.6).pdf(2)
```

## Poisson and binomial comparison

```python
RV(Binomial(100, 0.02)).sim(1000000).plot(jitter = True)
RV(Poisson(100 * 0.02)).sim(1000000).plot(jitter = True)
```

## Soccer problem 

```python
X, Y = RV(Poisson(2.2) * Poisson (1.8))

X.sim(10000).plot()
```

```python
Z = X + Y
(X | (Z == 3) ).sim(10000).plot()
```
