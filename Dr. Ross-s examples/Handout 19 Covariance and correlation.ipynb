# Ex 19.3

```python
def uniform_binomial():
    x = Uniform(a = 0, b = 1).draw()
    y = Binomial(n = 10, p = x).draw() 
    return x, y/10

X, Y = RV(ProbabilitySpace(uniform_binomial))

xy = (X & Y).sim(10000)
xy.plot(jitter = True, alpha = 0.1)
xy.cov(), xy.corr()
```

## Ex 19.5

```python
die = [1, 2, 3, 4, 5, 6]
P = BoxModel(die, size = 10)
W = RV(P, sum)
X = RV(P, count_eq(1))
Y = RV(P, count_eq(6))
Z = X + Y

(W & X & Y & Z).sim(10000).corr()
```

```python
(W & X).sim(1000).plot(jitter = True)
```

```python
(W & Y).sim(1000).plot(jitter = True)
```

```python
(W & Z).sim(1000).plot(jitter = True)
```

```python
(X & Y).sim(1000).plot(jitter = True)
```

## Ex 19.6 (assuming bivariate normal, but mean and SD calcs work in general)

```python
X, Y = RV(BivariateNormal(mean1 = 125, mean2 = 174, sd1 = 75, sd2 = 41, corr = -0.55))
(X & Y).sim(10000).plot(alpha = 0.1)
```

```python
z = (X + Y).sim(10000)
z.plot()
z.mean(), z.sd()
```
