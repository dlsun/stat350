# Handout 22 - Mean and Autocovariance functions, Stationary processes

```python
from symbulate import *
%matplotlib inline
```

## Exer 22.1

```python
def f(outcome, time):
    return 4 + outcome * cos(2 * pi * time)

P = Uniform(-1, 1)

X = RandomProcess(P, TimeIndex(fs = inf), f)
```

```python
x = X.sim(100)
x.plot(tmax = 3)
```

```python
x.plot(tmax = 3)
x.mean().plot('r')
```

```python
x.var().plot()
```

## Exer 22.3

```python
def f(outcome, time):
    return 4 + outcome * cos(2 * pi * time)

P = Uniform(-1, 1)

X = RandomProcess(P, TimeIndex(fs = inf), f)
```

```python
xy = (X[0] & X[1]).sim(1000)
xy.plot()
xy.cov()
```

```python
xy = (X[0] & X[0.5]).sim(1000)
xy.plot()
xy.cov()
```

```python
xy = (X[0] & X[0.25]).sim(1000)
xy.plot()
xy.cov()
```

```python
xy = (X[0] & X[0.1]).sim(1000)
xy.plot()
xy.cov()
```

## Exer 22.4

```python
def f(outcome, time):
    return cos(2 * pi * time  + outcome)

P = Uniform(-pi, pi)

X = RandomProcess(P, TimeIndex(fs=inf), f)
```

```python
X.sim(1).plot(alpha=1, tmax=3)
```

```python
X.sim(100).plot(tmax=3)
```

```python
x = X.sim(1000)
x.plot(alpha=0.01, tmax = 3)
x.mean().plot('r')
```

```python
x.var().plot('r')
```

```python
xy = (X[1] & X[1.25]).sim(1000)
xy.plot(jitter = True)
xy.cov(), xy.corr()
```

```python
xy = (X[1] & X[1.5]).sim(1000)
xy.plot(jitter = True)
xy.cov(), xy.corr()
```

```python
xy = (X[1] & X[2]).sim(1000)
xy.plot(jitter = True)
xy.cov(), xy.corr()
```

```python
xy = (X[1] & X[9/8]).sim(1000)
xy.plot(jitter = True)
xy.cov(), xy.corr()
```
