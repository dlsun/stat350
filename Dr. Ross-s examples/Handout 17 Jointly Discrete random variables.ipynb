# Handout 17 - Joint and conditional distributions (discrete)

```python
from symbulate import *
%matplotlib inline
```

# Exer 17.1

```python
P = BoxModel([1, 2, 3, 4], size=2)
X = RV(P, sum)
Y = RV(P, max)
```

```python
xy = (X & Y).sim(10000)
xy
```

```python
xy.tabulate(normalize=True)
```

```python
xy.plot()
```

```python
xy.plot(jitter=True)
```

```python
xy.plot("tile")
```

```python
xy.plot(["tile", "marginal"])
```

```python
y = Y.sim(10000)
y.plot()
y.mean(), y.sd()
```

```python
y_given_X_eq_6 = (Y | (X == 6) ).sim(10000)
y_given_X_eq_6.plot()
y_given_X_eq_6.tabulate()
```

```python
y_given_X_eq_6.mean(), y_given_X_eq_6.sd()
```

```python
area = (X * Y).sim(10000)
area.plot()
area.tabulate(normalize=True)
```

```python
area.mean()
```
