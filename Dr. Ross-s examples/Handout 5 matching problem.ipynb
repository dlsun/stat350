```python
from symbulate import *
%matplotlib inline
```

```python
# matching problem - outcomes
n = 4
labels = list(range(n))
P = BoxModel(labels, size = n, replace = False)
nsim = 10000
shuffles = P.sim(nsim)
shuffles
```

```python
# matching problem - number of matches
def number_matches(x):
    count = 0
    for i in range(0, n, 1):
        if x[i] == labels[i]:
            count += 1
    return count

number_matches((0,1,3,2))
```

```python
X = RV(P, number_matches)
values = X.sim(nsim)
values
```

```python
values.tabulate(normalize=True)
```

```python
values.plot()
```
