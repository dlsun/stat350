```python
N = PoissonProcess(rate = 2)
```

```python
N.sim(1).plot(alpha = 1)
```

```python
N.sim(100).plot()
N.sim(100).mean().plot()
```

```python
n = N[5].sim(10000)
n.plot()
Poisson(2 * 5).plot()
n.count_eq(8) / 10000
```
