## Exercise 8.1

```python
P = BoxModel([1, 0], probs=[2/3, 1/3], size=5, replace=True)
P.sim(5)
```

```python
X=RV(P, sum)
x = X.sim(10000)
x
```

```python
x.tabulate(normalize=True)
```

```python
x.plot()
```

```python
x.count_eq(5) / 10000
```

```python
1 - x.count_eq(0) / 10000
```

```python
P.sim(10000).count_eq((1, 1, 1, 1, 0)) / 10000
```

```python
x.count_eq(4) / 10000
```

```python
x.count_eq(3) / 10000
```

```python
x.plot()
Binomial(n=5, p=2/3).plot()
```

```python
Y = RV(Binomial(n=5, p=2/3))
Y.sim(10000).plot()
```
