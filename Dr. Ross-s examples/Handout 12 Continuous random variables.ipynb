## Handout 12 - Continuous RVs

```python
from symbulate import *
%matplotlib inline
```

```python
P = BoxModel(list(range(0, 9 + 1)), size=4, replace=True)
X = RV(P)
U = X[0]/10 + X[1]/100 + X[2]/1000 + X[3]/10000
```

```python
u = U.sim(50)
u.plot("rug")
```

```python
u = U.sim(10000)
```

```python
u.plot(bins=10)
```

```python
u.plot()
```

```python
U = RV(Uniform(0, 1))
U.sim(100000).plot()
```

```python
U = RV(Uniform(0, 1))
U.sim(100000).plot(bins=100)
```

```python
U = RV(Uniform(0, 1))
U.sim(1000000).plot(bins=100)
Uniform(0, 1).plot()
```

```python
a = 50
b = 100
U = RV(Uniform(0, 1))
X = a + (b - a) * U
X.sim(10000).plot()
Uniform(a, b).plot()
```

```python
U = RV(Uniform(0, 1))
X = -log(U)
X.sim(10000).plot()
Exponential(rate=1).plot()
```
