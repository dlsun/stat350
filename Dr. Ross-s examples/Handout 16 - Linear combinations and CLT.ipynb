# Exer 16.1

# a)

```python
population = BoxModel([10, 70, 200], probs=[0.4, 0.4, 0.2])
```

```python
X = RV(population)
x = X.sim(10000)
x.plot()
x.mean(), x.var(), x.sd()
```

# b)

```python
n = 2
X = RV(population ** n, mean)
x = X.sim(10000)
x.plot()
x.mean(), x.var(), x.sd()
```

# Exer 16.2

```python
X, Y = RV(Uniform(0, 1) ** 2)
S = X + Y
s = S.sim(10000)
s.plot()
s.mean(), s.var(), s.sd()
```
