```python
# Always run these two lines first
from symbulate import *
%matplotlib inline
```

```python
# equate True to fail and False to not to us Python logical functions
# probability model for 4 independent components
# each component fails with probability 0.1
p = 0.1
P = BoxModel([True, False], size = 1, probs = [p, 1-p])**4
nrep = 100000
component_fails = P.sim(nrep)
component_fails
```

```python
# parallel - system fails if all components fail 
parallel_fails = component_fails.apply(all)
parallel_fails
```

```python
parallel_fails.tabulate(normalize=True)
```

```python
# series - system fails if any components fails 
series_fails = component_fails.apply(any)
series_fails
```

```python
series_fails.tabulate(normalize=True)
```
