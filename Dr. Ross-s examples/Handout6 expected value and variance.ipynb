```python
from symbulate import *
%matplotlib inline
```

```python
P = BoxModel([1, -1], size=2)
X = RV(P, sum)

x = X.sim(10000)
x.plot()
x.mean()
```

```python
x.tabulate(normalize=True)
```

```python
Y = X**2
y = Y.sim(10000)
y.plot()
y.mean()
```

```python
X = RV(BoxModel([0, 3], probs=[0.05, 0.95]))
x = X.sim(10000)
x
```

```python
x.plot()
x.tabulate(normalize=True)
```

```python
x.mean()
```

```python
x.var(), x.sd()
```
