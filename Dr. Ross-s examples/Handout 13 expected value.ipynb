# Exer 13.2

### It's actually not super easy to simulate values from the pdf in Exercise 13.2.  Here is one way which uses "rejection sampling".  (You don't need to know this.)

```python
def f(x):
    return (1 / 9) * (4 - x **2)

U, Y = RV(Uniform(0, 1) * Uniform(-1, 2))

x = (Y | (U < (f(Y) / (4 /9)) ) ).sim(10000)
x.plot()
x.mean(), x.sd()
```
