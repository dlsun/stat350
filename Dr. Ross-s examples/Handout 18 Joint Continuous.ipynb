# Exer 18.1

```python
U, V = RV(Uniform(0, 1) ** 2)
X = -log(U)
Y = -log(V) / X

xy = (X & Y).sim(10000)
xy
```

```python
xy.plot()
```

```python
def zoom(xy):
    if (xy[1] < 50) & (xy[0] < 8):
        return True
    
xy = xy.filter(zoom)
xy.plot()
```

```python
xy.plot("hist", bins=[10,10])
```

```python
xy.plot("hist")
```

```python
xy.plot("density")
```

```python
def x1y2(x):
    return ((x[0] > 1) & (x[1] > 2))

len((X & Y).sim(10000).filter(x1y2)) / 10000
```

```python
X.sim(10000).plot()
Exponential(rate=1).plot()
```

```python
Y.sim(10000).plot()
```

```python
xy.plot(["hist", "marginal"])
```

```python
(Y | (X > 1) ).sim(10000).count_gt(2) / 10000
```

```python
(Y | (abs(X - 1/2) < 0.01) ).sim(10000).plot()
Exponential(rate=1/2).plot()
```

```python
(Y | (abs(X - 1/2) < 0.01) ).sim(10000).count_gt(2) / 10000
```

# Exer 18.2

```python
P = Uniform(0, 1) ** 2
X = RV(P, min)
Y = RV(P, max)
(X & Y).sim(1000).plot()
```

```python
xy = (X & Y).sim(10000)
xy.plot("hist")
```

```python
xy.plot("density")
```

```python
(X + Y).sim(10000).count_lt(1) / 10000
```

```python
x = X.sim(10000)
x.plot()
x.mean(), x.sd(), x.count_lt(1/3) / 10000
```

```python
xy.plot(["hist", "marginal"])
```

```python
x_given_Y_eq_onehalf = (X | (abs(Y - 1/2) < 0.01) ).sim(10000)
x_given_Y_eq_onehalf.plot()
x_given_Y_eq_onehalf.mean(), x_given_Y_eq_onehalf.sd(), x_given_Y_eq_onehalf.count_lt(1/3) / 10000
```

```python
product = (X * Y).sim(10000)
product.plot()
product.mean()
```

# Ex 18.3

```python
def exp_sim():
    x = Exponential(rate = 1/3).draw()
    y = Exponential(rate = 1/x).draw()
    return x, y

X, Y = RV(ProbabilitySpace(exp_sim))

(X & Y).sim(1000).plot()
```

```python
xy = (X & Y).sim(10000)
xy.plot("hist")
```

```python
xy = (X & Y).sim(10000)
xy.plot("density")
```

```python
xy = (X * Y).sim(10000)
xy.plot()
```

```python
xy.mean()
```

# Ex 18.4

```python
X, Y = RV(Exponential(rate = 2) * Normal(mean = 0, sd = 1))
xy = (X & Y).sim(10000)
xy.plot()
```

```python
xy.plot("density")
```

```python
xy.plot(["hist", "marginal"])
```

```python
x = X.sim(10000)
x.plot()
x.mean(), x.sd()
```

```python
x = (X | (abs(Y - 1) < 0.01) ).sim(1000)
x.plot()
x.mean(), x.sd()
```

```python
xy = (X * Y).sim(10000)
xy.plot()
xy.mean()
```
